```python
import copy 

def read_data(use_input = False):
    data = ""
    fname = "day4.input" if use_input else "day4.example"
    with open(fname, "r") as f:
        data = f.read()

    return list(map(
        lambda x: list(x),
        data.split("\n")
    ))

data = read_data(True)

empty_data = []
for i in range(len(data)):
    empty_data.append([])
    for j in range(len(data[0])):
        empty_data[-1].append("")

empty_data2 = copy.copy(empty_data)
```

```python

dritto = [[0,0], [0,1], [0,2], [0,3]]
sversa = [[0,0], [0,-1], [0,-2], [0,-3]]

top2bot = [[0,0], [1,0], [2,0], [3,0]]
bot2top = [[0,0], [-1,0], [-2,0], [-3,0]]

ftl2br = [[0,0], [1,1], [2,2], [3,3] ]
fbr2tl = [[0,0], [-1,-1], [-2,-2], [-3,-3] ]

fbl2tr = [[0,0], [-1,1], [-2,2], [-3,3]]
ftr2bl = [[0,0], [1,-1], [2,-2], [3,-3]]

cell_to_read =[
    dritto, sversa,
    top2bot, bot2top,
    ftl2br, fbr2tl,
    fbl2tr, ftr2bl
]
```

```python
import time
from IPython.display import clear_output
import copy

# data -> matrice appena letta
# current_pos -> (i,j)
# mask -> uno di cell_to_read
def read_mask(data, current_pos, mask):
    res = []
    i, j = current_pos
    data_max_x = len(data)
    data_max_y = len(data[0])
    for dx, dy in mask:
        new_x = i+dx
        new_y = j+dy
        if not ( 0<= new_x < data_max_x ): return None, True
        if not ( 0<= new_y < data_max_y ): return None, True

        res.append( data[new_x][new_y] )

    return res, False

def write_mask(data, current_pos, mask, character='.'):
    i, j = current_pos
    for dx, dy in mask:
        new_x = i+dx
        new_y = j+dy
        data[new_x][new_y] = character
    return data


def copy_data_where_not_empty(data, empty_data, character='.'):
    new_data = []
    for i in range(len(data)):
        new_data.append([])
        for j in range(len(data[0])):
            if empty_data[i][j] == character:
                new_data[-1].append(data[i][j])
            else:
                new_data[-1].append(character)
    return new_data



target = "XMAS"
target_count = 0

for i in range(len(data)):
    for j in range(len(data[0])):
        for m in cell_to_read:
            res, overflow = read_mask(copy.copy(data), (i,j), m)
            if overflow: 
                continue
            
            if "".join(res) == target:
                empty_data = write_mask(empty_data, (i,j), m)
                target_count += 1


            


print(f"{target_count=}")
```

```python
croceDX = (((-1,-1), (0,0), (1,1)), 
           ((-1,1), (0,0), (1,-1)))
croceDW = (((-1,-1), (0,0), (1,1)), 
           ((1,-1), (0,0), (-1,1)))
croceSX = (((1,-1), (0,0), (-1,1)),
            ((1,1), (0,0), (-1,-1)))
croceUP = (((-1,1), (0,0), (1,-1)),
            ((1,1), (0,0), (-1,-1)))

croci = [
    croceDX, croceUP, croceSX, croceDW
]

target_count = 0
for i in range(len(data)):
    for j in range(len(data[0])):
        for m in croci:
            res0, overflow = read_mask(copy.copy(data), (i,j), m[0])
            if overflow: 
                continue
            res1, overflow = read_mask(copy.copy(data), (i,j), m[1])
            if overflow: 
                continue
            
            if "".join(res0) == "MAS" and "".join(res1) == "MAS" :
                empty_data2 = write_mask(empty_data2, (i,j), m[0])
                empty_data2 = write_mask(empty_data2, (i,j), m[1])
                target_count += 1

print(f"{target_count=}")
```
